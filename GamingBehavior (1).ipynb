# 1. Questionnaire A

The participants had to rank themselfes on 7 statements about their psycological well-being:

1. Feeling nervous, anxious, or on edge
2. Not being able to stop or control worrying
3. Worrying too much about different things
4. Trouble relaxing
5. Being so restless that it's hard to sit still
6. Becoming easily annoyed or irritable
7. Feeling afraid as if something awful might happen

Each Statement could be answered with the following:

0 - Not at all
1 - Several days
2 - Over half the days
3 - Nearly every day

Lets plot a **spyder plot** and a **heatmap** for each statement, grouped by age and hours played:

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Assuming 'legit_hour_df' is your original dataframe
legit_hours_copy = legit_hour_df.copy()

# Statements mapped to their corresponding column names
statements = {
    'GAD1': 'Feeling nervous, anxious, or on edge',
    'GAD2': 'Not being able to stop or control worrying',
    'GAD3': 'Worrying too much about different things',
    'GAD4': 'Trouble relaxing',
    'GAD5': 'Being so restless that it`s hard to sit still',
    'GAD6': 'Becoming easily annoyed or irritable',
    'GAD7': 'Feeling afraid as if something awful might happen'
}

statement_columns = list(statements.keys())
statement_labels = list(statements.values())

# Creating bins for age
age_bins = [18, 20, 25, 30, 100]
age_labels = ['18-20', '20-25', '25-30', '30+']
legit_hours_copy['AgeBin'] = pd.cut(legit_hours_copy.Age, age_bins, labels=age_labels)

# Creating bins for hours played per week
hour_bins = [0, 20, 40, 60, 80, 120]
hour_labels = ['0-20', '20-40', '40-60', '60-80', '80+']
legit_hours_copy['HoursBin'] = pd.cut(legit_hours_copy.Hours, hour_bins, labels=hour_labels)

# Prepare data for radar charts
def prepare_radar_data(df):
    radar_data = df.groupby(['AgeBin', 'HoursBin'])[statement_columns].mean().reset_index()
    return radar_data

radar_data = prepare_radar_data(legit_hours_copy)

# Create subplots for each age zone
fig = make_subplots(rows=len(age_labels), cols=1,
                    subplot_titles=age_labels,
                    specs=[[{'type': 'polar'}] for _ in age_labels])

# Iterate over each age zone and add radar plot
for idx, age_zone in enumerate(age_labels):
    age_data = radar_data[radar_data['AgeBin'] == age_zone]
    for _, row in age_data.iterrows():
        fig.add_trace(go.Scatterpolar(
            r=row[statement_columns].values,
            theta=statement_labels,
            fill='toself',
            name=f"{row['AgeBin']} | {row['HoursBin']}"
        ), row=idx+1, col=1)

fig.update_layout(
    polar=dict(
        radialaxis=dict(visible=True, showticklabels=False)
    ),
    showlegend=True,
    title="Radar Chart for Psychological Well-being by Age and Hours Played",
    height=2000,  # Set the height of the entire figure
    width=1000    # Set the width of the entire figure
)
fig.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Assuming 'legit_hour_df' is your original dataframe
legit_hours_copy = legit_hour_df.copy()

# Affection dictionary
affection_dict = {
    'Not difficult at all': 0,
    'Somewhat difficult': 1,
    'Very difficult': 2,
    'Extremely difficult': 3,
}

# Statements mapped to their corresponding column names
statements = {
    'GAD1': 'Feeling nervous, anxious, or on edge',
    'GAD2': 'Not being able to stop or control worrying',
    'GAD3': 'Worrying too much about different things',
    'GAD4': 'Trouble relaxing',
    'GAD5': 'Being so restless that it`s hard to sit still',
    'GAD6': 'Becoming easily annoyed or irritable',
    'GAD7': 'Feeling afraid as if something awful might happen'
}

statement_columns = list(statements.keys())
statement_labels = list(statements.values())

# Replace text responses with numerical values
for col in statement_columns:
    if legit_hours_copy[col].dtype == 'object':
        legit_hours_copy[col] = legit_hours_copy[col].map(affection_dict)
        print(f"Column {col} after mapping:")
        print(legit_hours_copy[col].value_counts(dropna=False))
    else:
        print(f"Column {col} is not of type object and might already be numeric")

# Drop rows where all statement columns are NaN
legit_hours_copy.dropna(subset=statement_columns, how='all', inplace=True)

# Creating bins for age
age_bins = [18, 20, 25, 30, 100]
age_labels = ['18-20', '20-25', '25-30', '30+']
legit_hours_copy['AgeBin'] = pd.cut(legit_hours_copy['Age'], bins=age_bins, labels=age_labels)

# Creating bins for hours played per week
hour_bins = [0, 20, 40, 60, 80, 120]
hour_labels = ['0-20', '20-40', '40-60', '60-80', '80+']
legit_hours_copy['HoursBin'] = pd.cut(legit_hours_copy['Hours'], bins=hour_bins, labels=hour_labels)

# Prepare data for heatmap
heatmap_data = legit_hours_copy.groupby(['AgeBin', 'HoursBin'])[statement_columns].mean().reset_index()

# Debug: Print heatmap data to ensure it's correctly processed
print("Heatmap Data:")
print(heatmap_data)

# Create heatmap for each statement
fig = make_subplots(
    rows=len(statement_columns), cols=1,
    subplot_titles=statement_labels,
    vertical_spacing=0.1,
    specs=[[{'type': 'heatmap'}] for _ in statement_columns]
)

# Iterate over each statement to add heatmap
for idx, statement in enumerate(statement_columns):
    heatmap_values = heatmap_data.pivot(index='AgeBin', columns='HoursBin', values=statement)
    
    # Debug: Print heatmap values for each statement
    print(f"Heatmap values for {statement}:")
    print(heatmap_values)
    
    fig.add_trace(go.Heatmap(
    z=heatmap_values.values if not heatmap_values.empty else np.zeros((len(age_labels), len(hour_labels))),
    x=heatmap_values.columns if not heatmap_values.empty else hour_labels,
    y=heatmap_values.index if not heatmap_values.empty else age_labels,
    colorscale='Viridis',
    colorbar=dict(title='Mean Score'),
    text=heatmap_values.values if not heatmap_values.empty else np.zeros((len(age_labels), len(hour_labels))),
    texttemplate="%{text:.2f}",  # Format to display two decimal places
    textfont={"size": 10}
), row=idx+1, col=1)

# Update layout
fig.update_layout(
    height=7000,  # Increase the height of the figure
    width=1000,   # Increase the width of the figure
    title="Heatmap of Psychological Well-being by Age and Hours Played",
)

fig.show()
```

**Younger Age Groups (18-25)**: Show higher levels of anxiety across various factors, especially related to restlessness, irritability, and nervousness. This suggests a need for targeted interventions to help manage anxiety for younger gamers.

**Mid Age Group (25-30)**: Experiences moderate anxiety levels with some high points in restlessness and irritability, indicating a transition phase where anxiety starts to decrease but still requires attention.

**Older Age Group (30+)**: Exhibits the lowest anxiety levels, indicating better management of psychological well-being and reduced impact of anxiety factors.

# Affects of these problems on daily live

-The participants were asked "...how difficult have these made it for you to do your work, take care of things at home, or get along with other people?"

-They had the following choice of answers:

  0 - Not difficult at all
  1 - Somewhat difficult
  2 - Very difficult
  4 - Extremely difficult
-note: the integers before the answers are my representation of them in the dataframe, used for calculating the mean.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'legit_hour_df' is your original dataframe
legit_hours_copy = legit_hour_df.copy()

# Affection dictionary
affection_dict = {
    'Not difficult at all': 0,
    'Somewhat difficult': 1,
    'Very difficult': 2,
    'Extremely difficult': 3,
}

# Translate GADE labels to numeric representation for mean calculation
if 'GADE' in legit_hours_copy.columns:
    legit_hours_copy['GADE_numeric'] = legit_hours_copy['GADE'].map(affection_dict)
    print(f"GADE column after mapping:")
    print(legit_hours_copy['GADE_numeric'].value_counts(dropna=False))
else:
    raise KeyError("GADE column not found in the dataframe")

# Ensure AgeBin and HoursBin are properly created in the dataframe
# Creating bins for age
age_bins = [18, 20, 25, 30, 100]
age_labels = ['18-20', '20-25', '25-30', '30+']
legit_hours_copy['AgeBin'] = pd.cut(legit_hours_copy['Age'], bins=age_bins, labels=age_labels)

# Creating bins for hours played per week
hour_bins = [0, 20, 40, 60, 80, 120]
hour_labels = ['0-20', '20-40', '40-60', '60-80', '80+']
legit_hours_copy['HoursBin'] = pd.cut(legit_hours_copy['Hours'], bins=hour_bins, labels=hour_labels)

# Function to plot heatmap for a given statement
def plot_heatmap_for_statement(df, current_stmt, col, min_val, max_val):
    # Pivot table to create the heatmap data
    heatmap_data = df.pivot_table(index='AgeBin', columns='HoursBin', values=col, aggfunc='mean')
    
    # Fill missing values with a placeholder (e.g., 0)
    heatmap_data_filled = heatmap_data.reindex(index=age_labels, columns=hour_labels, fill_value=0)

    plt.figure(figsize=(10, 8))
    sns.heatmap(heatmap_data_filled, annot=True, cmap='viridis', vmin=min_val, vmax=max_val, cbar_kws={'label': 'Mean Score'})
    plt.title(f'Heatmap of {current_stmt}')
    plt.xlabel('Hours Played per Week')
    plt.ylabel('Age Group')
    plt.show()

# Plot heatmap for GADE_numeric
plot_heatmap_for_statement(df=legit_hours_copy, current_stmt='Affect on daily live', col='GADE_numeric', min_val=0, max_val=3)
```

**A similar pattern as above can be seen, where people who play more, tend to have bigger affects of the above problems in their lifes.**
